```python
def floor_abs(x,fl=10**(-6)):
    idxs_good = np.nonzero(abs(x)<fl)
    idxs_high = np.nonzero(x>=fl)
    idxs_low = np.nonzero(x<=-fl)
    y = np.zeros(len(x))
    y[idxs_good] = x[idxs_good]
    y[idxs_high] = np.ones(len(idxs_high)) * fl
    y[idxs_low] = - np.ones(len(idxs_low)) * fl
    return y
```
